Encoding can be done using different way two of them are done below:
1. Label Encoding
2. Ordinal Encoding

**Label Encoding**

Label Encoding is done on Nominal data that donot follow a sequence like names of different animals (cow, cat, dog, elephant, etc.)

```python
dataset = pd.read_csv('../teams.csv')
dataset.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dataset.columns
```

```python
dataset['coaches']
```

```python
dataset['coaches_en'] = dataset['coaches'].fillna(dataset['coaches'].mode()[0], inplace=True)
```

```python
dataset.tail(10)
```

```python
dataset['coaches']=le.fit_transform(dataset['coaches'])
```

```python
dataset['coaches']
```

The label encoding on nominal data is not reccommneded. Data Scientists use Ordinal encoding mostly.


**Ordinal Encoding**

Ordinal Encoding is done using two ways:
1. Using sklearn library
2. Map function

```python
df = pd.DataFrame({'size': ['s','m','x','s','m','xl','m','x','s','s','m','xl','m','m','x','s',]})
df.head(5)
```

```python
ord_data = [['s','m','l','x','xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=ord_data)
```

```python
oe.fit(df[['size']])
```

```python
df['en_size'] = oe.transform(df[['size']])
```

```python
df
```

Now performing ordinal encoding with map function

```python
df1 = pd.DataFrame({'size': ['s','m','x','s','m','xl','m','x','s','s','m','xl','m','m','x','s',]})
```

```python
ord_data_new = {'s':0, 'm':1,'x':2, 'xl':3}
```

```python
df1['size'].map(ord_data_new)
```

```python
df1['size_en'] = df1['size'].map(ord_data_new)
```
